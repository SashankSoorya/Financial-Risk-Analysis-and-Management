```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd

# Define tickers and date range
tickers = ['J&KBANK.NS', 'JAICORPLTD.NS', 'JAYNECOIND.NS', 'JCHAC.NS', '^NSEI']
start_date = '2020-04-01'
end_date = '2024-09-30'

# Initialize an empty DataFrame to store the stock data
df = pd.DataFrame()

# Download stock data for each ticker and add to the DataFrame
for ticker in tickers:
    tick = yf.Ticker(ticker)
    stock_data = tick.history(start=start_date, end=end_date)['Close']
    df[ticker] = stock_data

# Drop rows with missing values (if any)
df.dropna(inplace=True)

# Calculate daily returns
daily_returns = df.pct_change()

# Calculate weekly returns
weekly_returns = df.resample('W').last().pct_change()

# Calculate monthly returns
monthly_returns = df.resample('MS').last().pct_change()

# Analyze returns and plot
for ticker in tickers:
    # Calculate statistics
    daily_mean = daily_returns[ticker].mean()
    daily_std = daily_returns[ticker].std()
    weekly_mean = weekly_returns[ticker].mean()
    weekly_std = weekly_returns[ticker].std()
    monthly_mean = monthly_returns[ticker].mean()
    monthly_std = monthly_returns[ticker].std()

    # Print statistics
    print(f"=== Statistics for {ticker} ===")
    print(f"Daily Returns: Mean = {daily_mean:.6f}, Std Dev = {daily_std:.6f}")
    print(f"Weekly Returns: Mean = {weekly_mean:.6f}, Std Dev = {weekly_std:.6f}")
    print(f"Monthly Returns: Mean = {monthly_mean:.6f}, Std Dev = {monthly_std:.6f}")
    print()

    # Plot daily returns
    plt.figure(figsize=(10, 6))
    plt.plot(daily_returns.index, daily_returns[ticker], label=f'{ticker} Daily Returns', color='blue', alpha=0.7)
    plt.title(f'Daily Returns of {ticker} from {start_date} to {end_date}')
    plt.xlabel('Date')
    plt.ylabel('Returns')
    plt.legend()
    plt.show()

    # Plot weekly returns
    plt.figure(figsize=(10, 6))
    plt.plot(weekly_returns.index, weekly_returns[ticker], label=f'{ticker} Weekly Returns', color='green', alpha=0.7)
    plt.title(f'Weekly Returns of {ticker} from {start_date} to {end_date}')
    plt.xlabel('Date')
    plt.ylabel('Returns')
    plt.legend()
    plt.show()

    # Plot monthly returns
    plt.figure(figsize=(10, 6))
    plt.plot(monthly_returns.index, monthly_returns[ticker], label=f'{ticker} Monthly Returns', color='red', alpha=0.7)
    plt.title(f'Monthly Returns of {ticker} from {start_date} to {end_date}')
    plt.xlabel('Date')
    plt.ylabel('Returns')
    plt.legend()
    plt.show()
```
